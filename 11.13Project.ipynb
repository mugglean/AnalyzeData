```python
def getTable(User_id):
    df=pd.read_csv("F:\\BWMProject\\11.5\\Sorted_table.csv")
    User_df=df.loc[df.User_id==User_id]
    User_df.drop(['Unnamed: 0','Venue_id','Time_offset','Venue_category_name','Country'],axis=1,inplace=True)
    Sr=pd.to_datetime(User_df.UTC_time)
    #拿一个List存储
    dateList=[]
    for key,value in Sr.items():
        dateList.append(value)
    
    
    # 尝试使用while1进行迭代
    alist=[]
    tIndex=0
    inintIndex=0
    finalIndex=0
    while True:
    #     找到下一个Index
        for i in range(inintIndex,len(dateList)):
            timedeltaour=(dateList[i]-dateList[inintIndex]).total_seconds()/3600
            if timedeltaour > 6:
                countNextIndex=i
                tIndex+=1
                break
    # 把分组信息添加到list
        strNmae="t"+str(tIndex)
        for i in range(inintIndex,countNextIndex):
            alist.append(strNmae)
        inintIndex=countNextIndex
    #     if countNextIndex==(len(dateList)-1):
    #     if countNextIndex==(len(dateList)-2) or countNextIndex>=len(dateList):
        if countNextIndex==(len(dateList)-1):
            tIndex+=1
            strNmae="t"+str(tIndex)
            alist.append(strNmae)
            break
    User_df['Time_Interval']=alist
    User_df.to_csv("F:\\BWMProject\\11.13\\"+str(User_id)+".csv",index = False)
```
