```python
import pandas as pd
import numpy as np
```

```python
sum_location_table=pd.read_csv("F:\\BWMProject\\10.2\\locationdict_dict_final.csv")
sum_location_table
```

```python
sum_table=pd.read_csv("F:\\BWMProject\\10.1\\result_filter.csv")
sum_table
```

输出两个表要merge的location列，发现并不是一样的，它们相差了几位小数

```python
print(sum_location_table.loc[0,'Latitude'])
print(sum_table.loc[0,'Latitude'])
```

使用apply进行取位置操作，以便得到相同的点，才能进行merge

```python
sum_table['Longitude']=sum_table['Longitude'].apply(lambda v :round(v, 6))
sum_location_table['Longitude']=sum_location_table['Longitude'].apply(lambda v :round(v, 6))
```

```python
print(sum_location_table.loc[0,'Longitude'])
print(sum_table.loc[0,'Longitude'])
```

```python
sum_table['Latitude']=sum_table['Latitude'].apply(lambda v :round(v, 6))
sum_location_table['Latitude']=sum_location_table['Latitude'].apply(lambda v :round(v, 6))
```

```python
print(sum_location_table.loc[0,'Latitude'])
print(sum_table.loc[0,'Latitude'])
```

进行Merge操作

```python
RES = pd.merge(sum_table,sum_location_table, on= ['Latitude','Longitude'] )
```
