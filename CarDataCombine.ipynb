```python
# 绘图所需要函数
# numList_seven：二维List
# timeList_seven:List，每个存储了日期
def plotWeekData(weekname,numList_seven,timeList_seven):
    for i in range(len(timeList_seven)):
        plt.plot(numList_seven[i],label=timeList_seven[i])
    plt.xlabel(u'Time',fontproperties='SimHei',fontsize=14)
    plt.ylabel(u'Num',fontproperties='SimHei',fontsize=14)
    plt.title(weekname)
    plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
    plt.xticks([0,48,96,144,192,242,290],['00:00','04:00','08:00','12:00','16:00','20:00','24:00'])
#     plt.savefig("F:\\BWMProject\\11.2\\"+weekname+".png")
    fig = plt.gcf()
    fig.set_size_inches(19, 10.5)
    fig.savefig("F:\\BWMProject\\11.2\\"+weekname+".png", dpi=100)
    plt.close()
```

```python
def getSeri(day):
    df=pd.read_csv("F:\\BWMProject\\taxi data\\"+day+".csv")
    cor=df.coordinates
    return cor
```

绘制第一周

```python
numList_seven=[]
timeList_seven=[]
day_index=0
week_str=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
for i in range(3,10):
    day="2018-09-"+str(0)+str(i) if i<10 else "2018-09-"+str(i)
    cor=getSeri(day)
    numList=[]
    for key,value in cor.items():
        strList=value.split(",")
        numList.append(int(len(strList)/2))
    numList_seven.append(numList)
    timeList_seven.append(day+week_str[day_index])
    day_index+=1
```

```python
plotWeekData("week1",numList_seven,timeList_seven)
```
