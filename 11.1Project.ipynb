```python
import pandas as pd
import numpy as np
from gmplot import gmplot
```

```python
df=pd.read_csv("F:\\BWMProject\\9.25\\result.csv")
```

```python
df
```

```python
df.groupby("User_id").size().sort_values()
```

```python
most_active_User=df.loc[df.User_id==73]
most_active_User
```

```python
cor_table=pd.DataFrame()
cor_table['Latitude']=most_active_User.Latitude
cor_table['Longitude']=most_active_User.Longitude
cor_table
```

```python
gmap_end = gmplot.GoogleMapPlotter(1.335946,103.906772,13)
gmap_end.scatter(cor_table['Latitude'], cor_table['Longitude'], color='red', size=40, marker=False)
gmap_end.draw("poi.html")
```
