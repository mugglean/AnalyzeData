先处理第一个文件，以便防止错误

```python
location_table_1=pd.read_csv("F:\\BWMProject\\10.2\\locationdict_1.csv")
location_table_1
```

```python
#使用API查询
APIKey='***********'
gmaps = googlemaps.Client(key=APIKey)
def getStreetInfo(x,y):
    alist=(x,y)
    reverse_geocode_result = gmaps.reverse_geocode(alist)
    try:
        return reverse_geocode_result[0]['formatted_address']
    except IndexError:
        return 0
```

```python
street_info = []
#第一个是np.int类型数据，第二个是Serilize型
for row_index,row in location_table_1.iterrows():
#    print (type(row['Latitude']),type(row['Longitude'])) 都是np.float
    return_sring=getStreetInfo(row['Latitude'],row['Longitude'])
    street_info.append(return_sring)
street_info
```

```python
location_table_1['street_info']=street_info
location_table_1
```

```python
location_table_1.to_csv("F:\\BWMProject\\10.2\\locationdict_1_result.csv",index=False)
```
