```python
from gmplot import gmplot
import numpy
import pandas as pd
```

```python
sum_table=pd.read_csv("F:\\BWMProject\\10.3\\final_table_Sin.csv")
sum_table
```

```python
sorted_table=sum_table.sort_values(by='User_id')
sorted_table=sorted_table.reset_index(drop=True)
sorted_table
```

é‡æŽ’User_id

```python
User_id_table=pd.DataFrame()
User_id_table["User_id"]=sorted_table.User_id
User_id_table
```

```python
User_id_table=User_id_table.drop_duplicates()
User_id_table=User_id_table.reset_index(drop=True)
User_id_table["new_User_id"]=range(1,User_id_table.shape[0]+1)
User_id_table
```

```python
resort_Userid_table=pd.merge(sorted_table,User_id_table,on='User_id')
resort_Userid_table["User_id"]=resort_Userid_table["new_User_id"]
del resort_Userid_table["new_User_id"]
resort_Userid_table
```

```python
User_id_table.to_csv("F:\\BWMProject\\10.8\\User_id_dict.csv",encoding='utf-8', index = False)
```

```python
resort_Userid_table.to_csv("F:\\BWMProject\\10.8\\resort_Userid_table.csv",encoding='utf-8', index = False)
```

é‡æŽ’POI_id

```python
POI_id_table=pd.DataFrame()
POI_id_table["Venue_category_name"]=resort_Userid_table.Venue_category_name
POI_id_table
```

```python
POI_id_table.drop_duplicates(inplace=True)
POI_id_table.reset_index(drop=True)
```

```python
POI_id_table["new_POI_id"]=range(1,POI_id_table.shape[0]+1)
POI_id_table
```

```python
resort_POIid_table=pd.merge(resort_Userid_table,POI_id_table,on='Venue_category_name')
resort_POIid_table
```

```python
del resort_POIid_table["Venue_category_name"]
```

```python
resort_POIid_table=resort_POIid_table.sort_values(by="User_id")
resort_POIid_table
```

```python
resort_POIid_table.to_csv("F:\\BWMProject\\10.8\\resort_POIid.csv",encoding='utf-8', index = False)
POI_id_table.to_csv("F:\\BWMProject\\10.8\\POI_id_dict.csv",encoding='utf-8', index = False)
```

é‡æŽ’Street_id

```python
resort_Street_id=pd.DataFrame()
resort_Street_id["street_info"]=resort_POIid_table.street_info
```

```python
resort_Street_id.drop_duplicates(inplace=True)
resort_Street_id.reset_index(drop=True)
```

```python
resort_Street_id["new_street_id"]=range(1,resort_Street_id.shape[0]+1)
resort_Street_id
```

```python
resort_Streetid_table=pd.merge(resort_POIid_table,resort_Street_id,on='street_info')
```

```python
del resort_Streetid_table["street_info"]
```

```python
resort_Streetid_table=resort_Streetid_table.sort_values(by=['User_id',"new_POI_id","new_street_id"])
resort_Streetid_table
```

```python
resort_Streetid_table.to_csv("F:\\BWMProject\\10.8\\resort_Streetid.csv",encoding='utf-8', index = False)
resort_Street_id.to_csv("F:\\BWMProject\\10.8\\street_id_dict.csv",encoding='utf-8', index = False)
```
